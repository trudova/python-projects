* df['column name'].dropna() - will give a new series with all na valus being droped

* df.dropna(how='all') drop only rows with all NA

* df.dropna(axis = 1) will drop all columns with missing values

#### if apply id applied to df , then the functionality of the custome func will be applyed 
to each column

```python
def get_fam_size(cell_value):
    fam_size= cell_value.sibsp + cell_value.parch
    if fam_size ==0:
        return 'solo'
    elif fam_size<5:
        return 'average'
    else:
        return 'large'
        
titanic['fam_size']=titanic.apply(get_fam_size, axis=1) # to apply on the row, the axis must be 1
```

## Combining series and data frames

#### pd.concat([series1, series2])
will stack 2 series on top of one another in the order they were provided 
index will be preserved from the original
to get new index :
* pd.concat([series1, series2], ignore_index=True)

#### pd.concat([series1, series2], axis=1) 
will join them side by side, forming two columns, and the indexis will be mached, if there is no match, cells will be filled in with , we can specify:
* join= 'inner' pd.concat([series1, series2], axis=1, join= 'inner')¶ give only overlaping index values

### works the same on df

## merge

df1.merge(df2) will look for overlaping colump and join then on the matching valus of that same column

* df1.merge(df2, on='name of your column or list of columns') in this case how is defalt inner
 * df1.merge(df2, on='name of your column or list of columns', how='left') will do left join , will join all from left df and all overlaps
 *  df1.merge(df2, on='name of your column or list of columns',  how='right') will do right join , will join all from right df and all overlaps
 *    df1.merge(df2, on='name of your column or list of columns',  how='outer') will join all of it
