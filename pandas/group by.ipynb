```python
shots= liga.groupby('Team').agg(
    total=('Shots','sum'),
    on_target=('Shots on target', 'sum')
)
shots
```

```python
shots['acuracy'] = (shots['on_target']/ shots['total']) * 100
```

```python
shots
```

```python
fig, ax = plt.subplots(2,1, figsize=(5,8), sharex=True)
shots['acuracy'].nlargest(10).sort_values(ascending=True).plot(kind="barh", ax=ax[0], title= "most scored")

shots['acuracy'].nsmallest(10).plot(kind="barh", ax=ax[1], title='most mised', color='red')
plt.xlabel('on target %')
```

```python
def get_age_group(age):
    if age <2:
        return 'infant'
    elif age <12:
        return 'child'
    elif age <20:
        return 'teen'
    elif age< 65:
        return 'adult'
    else:
        return 'senior'
    
```

```python
get_age_group(70)
```

```python
titanic['age_group']=titanic['age'].apply(get_age_group)
```

```python
def get_fam_size(s):
    fam_size = s.sibsp + s.parch
    if fam_size ==0:
        return 'solo'
    elif fam_size <5:
        return 'averege'
    else:
        return 'large'
```
